```python
import folium
```

# Pattern-of-life (PoL) Analysis

The purpose of this notebook is to demonstrate and visualize a case study in applying techniques for pattern-of-life analysis of a target given Find My location data. The specific case is that someone stole a pair of Airpods. Exact information on addresses and some routes taken, along with modes of transportation, are known, but they have been anonymized to comply with privacy laws.

### Limitations

- Find My does not provide historical location data; a user must persistently monitor the app to try and build this location history. Some locations or movements could then be missing.

### Interpreting the map

- Hover over an element to get a description.
- Click on an element to show the date or time indicating when the AirPods were at a location or when travel occurred.
- A dotted line indicates that the route is unknown. A solid line symbolizes that all route details are known.

## June 28

Theft occured on a train at approximately 11:30, most likely at Wien Hbf. The theft went unnoticed for about two hours before the Find My app was checked for the first time, where the AirPods were then located at a residential address in the 16th district (Ottakring). Therefore, the route taken by the AirPods to the address is unknown.

```python
m = folium.Map(location=(48.2085, 16.3721),
               zoom_start=12)

# --- location of theft ----
theft_coords = [48.1848, 16.3782]
folium.Marker(
    location=theft_coords,
    tooltip="Site of Theft - Wien Hbf",
    popup="~11:30 at June 28",
    icon=folium.Icon(color="red", icon="flag"),
).add_to(m)

# --- suspected residential address in Ottakring ---
ottakring_coords = [48.2128, 16.3178]
folium.Marker(
    location=ottakring_coords,
    tooltip="Suspected Residential Address",
    popup="June 28-29",
    icon=folium.Icon(color="orange", icon="home"),
).add_to(m)
# uncertainty radius
folium.Circle(
    location=ottakring_coords,
    radius=500,
    color="orange",
    weight=2,
    fill=True,
    fill_color="orange",
    fill_opacity=0.2
).add_to(m)

# --- add unknown route ---
folium.PolyLine(
    locations=[theft_coords, ottakring_coords],
    tooltip="Wien Hbf -> Ottakring",
    popup="Afternoon of June 28",
    color="red",
    weight=3,
    opacity=0.8,
    dash_array="5, 10"   # 5 px line, 10 px gap
).add_to(m)

m
```

## June 29-30

AirPods were observed for a short time in the morning along Thaliastraße near the residential address (not marked), and police did a quick search of the area with no results, which was not surprising. The AirPods spent the rest of the day at the residential address.

Movement was detected late at night, with the target moving from Ottakring to another residential address near the Naschmarkt. The airpods were taken out from the case for the first time, and I immediately executed the "Play Sound" command on both of them. The command might have been successful since no location data was reported for the next ~48 hours, suggesting that something was done to suppress location reports.

The exact route, timestamps, and most likely modes of transportation from the residential address in Ottakring to the Naschmarkt are known.

```python
# --- Naschmarkt address ---
naschmarkt_coords = [48.1986, 16.3612]
folium.Marker(
    location=naschmarkt_coords,
    tooltip="Airpods taken out of case for first time",
    popup="Around midnight of June 30",
    icon=folium.Icon(color="blue", icon="flash"),
).add_to(m)
# uncertainty radius
folium.Circle(
    location=naschmarkt_coords,
    radius=500,
    color="blue",
    weight=2,
    fill=True,
    fill_color="blue",
    fill_opacity=0.2
).add_to(m)

# --- add known route ---
folium.PolyLine(
    locations=[ottakring_coords, naschmarkt_coords],
    tooltip="Ottakring -> Naschmarkt",
    popup="Around midnight of June 30",
    color="red",
    weight=3,
    opacity=0.8,
).add_to(m)

m
```

## July 2

After ~48 hours of silence, the AirPods reported a new location around midnight at a suspected residential address in the 22nd district (Donaustadt). Due to the long period of silence and the drastic change in location, a handover may have occurred.

Interestingly the location was about an hour old but executing the "Play Sound" command immediately refreshed the location. Before, there was no feedback that a sound actually played for the case. In constrast, the AirPods tell you that the command is queued and will execute next time there's a connection to the AirPods.

In the morning, the Airpods travelled to Wien Meidling and boarded a train heading to Ukraine. The exact route, timestamps, and modes of transportation from the Donaustadt address to Wien Meidling is known.

```python
# --- suspected residential address in Donaustadt ---
donaustadt_coords = [48.2493, 16.4510]
folium.Marker(
    location=donaustadt_coords,
    tooltip="Suspected Residential Address",
    popup="July 2",
    icon=folium.Icon(color="orange", icon="home"),
).add_to(m)
# uncertainty radius
folium.Circle(
    location=donaustadt_coords,
    radius=500,
    color="orange",
    weight=2,
    fill=True,
    fill_color="orange",
    fill_opacity=0.2
).add_to(m)

# --- Naschmark -> Donaustadt ---
folium.PolyLine(
    locations=[naschmarkt_coords, donaustadt_coords],
    tooltip="Naschmark -> Donaustadt",
    popup="Between June 30 and July 2",
    color="red",
    weight=3,
    opacity=0.8,
    dash_array="5, 10"   # 5 px line, 10 px gap
).add_to(m)

# --- Wien Meidling ---
meidling_coords = [48.1745, 16.3341]
folium.Marker(
    location=meidling_coords,
    tooltip="Wien Meidling",
    popup="Morning of July 2",
    icon=folium.Icon(color="green", icon="train", prefix="fa"),
).add_to(m)

# --- Donaustadt -> Wien Meidling ---
folium.PolyLine(
    locations=[donaustadt_coords, meidling_coords],
    tooltip="Donaustadt -> Wien Meidling",
    popup="Morning of July 2",
    color="red",
    weight=3,
    opacity=0.8,
).add_to(m)

m
```
