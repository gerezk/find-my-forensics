# Train travel from Vienna to Ukraine

The notebook attempts to trace the route taken by the AirPods from Vienna, Austria to Kyiv, Ukraine (July 2-3). Exact information on addresses and some routes taken, along with modes of transportation, are known, but they have been anonymized to comply with privacy laws.

The train taken from Vienna, Austria to Chop, Ukraine and its schedule is known due to correlating Find My location reports to a live maps and timetables of trains in Hungary: https://mavplusz.hu/#/. The exact train taken from Chop to Kyiv could not be correlated to a train on the published timetables: https://www.uz.gov.ua/en/passengers/timetable/. A possible train was correlated with low confidence, but this was made on the assumption that it was severely delayed. No direct evidence supporting this assumption exists, but there was news reporting that a Russian air assault severely delayed trains in the east of the country the night of July 2: https://hmh.news/en/34839/due-to-shelling-of-kyiv-41-ukrainian-railways-trains-are-delayed/amp/.

### Limitations

- Find My does not provide historical location data; a user must persistently monitor the app to try and build this location history. Some locations or movements could then be missing.

### Interpreting the map

- Hover over an element to get a description.
- Click on an element to show the date or time indicating when the AirPods were at a location or when travel occurred.
- A dotted line indicates that the route details are not fully known. A solid line symbolizes that all route details are known.

```python
m = folium.Map(location=(49.2074, 23.1393),
               zoom_start=6)

folium.TileLayer(
    tiles="https://tiles.openrailwaymap.org/standard/{z}/{x}/{y}.png",
    attr='© OpenStreetMap contributors, OpenRailwayMap',
    name="OpenRailwayMap",
    overlay=False,
    control=True,
).add_to(m)

folium.LayerControl().add_to(m)
```

```python
# --- Wien Meidling ---
meidling_coords = [48.1745, 16.3341]
folium.Marker(
    location=meidling_coords,
    tooltip="Wien Meidling",
    popup="Morning of July 2",
    icon=folium.Icon(color="green", icon="train", prefix="fa"),
).add_to(m)

# --- Chop, Ukraine ---
chop_coords = [48.4324, 22.2055]
folium.Marker(
    location=chop_coords,
    tooltip="Chop, Ukraine",
    popup="Evening of July 2",
    icon=folium.Icon(color="green", icon="random", prefix="glyphicon"),
).add_to(m)
# --- Wien Meidling -> Chop, Ukraine ---
folium.PolyLine(
    locations=[meidling_coords, chop_coords],
    tooltip="Wien Meidling -> Chop, Ukraine",
    color="red",
    weight=3,
    opacity=0.8,
).add_to(m)

# --- Krasne, Ukraine ---
krasne_coords = [49.9179, 24.6134]
folium.Marker(
    location=krasne_coords,
    tooltip="Krasne, Ukraine",
    popup="Morning of July 3",
    icon=folium.Icon(color="green", icon="train", prefix="fa"),
).add_to(m)
# --- Chop, Ukraine -> Krasne, Ukraine ---
folium.PolyLine(
    locations=[chop_coords, krasne_coords],
    tooltip="Chop, Ukraine -> Krasne, Ukraine",
    color="red",
    weight=3,
    opacity=0.8,
    dash_array="5, 10"   # 5 px line, 10 px gap
).add_to(m)

# --- Sytne, Ukraine ---
sytne_coords = [50.1762, 25.4329]
folium.Marker(
    location=sytne_coords,
    tooltip="Sytne, Ukraine",
    popup="Morning of July 3",
    icon=folium.Icon(color="green", icon="train", prefix="fa"),
).add_to(m)
# --- Krasne, Ukraine -> Sytne, Ukraine ---
folium.PolyLine(
    locations=[krasne_coords, sytne_coords],
    tooltip="Krasne, Ukraine -> Sytne, Ukraine",
    color="red",
    weight=3,
    opacity=0.8,
    dash_array="5, 10"   # 5 px line, 10 px gap
).add_to(m)

# --- Staryi Kryvyn, Ukraine ---
staryi_kryvyn_coords = [50.3494, 26.7077]
folium.Marker(
    location=staryi_kryvyn_coords,
    tooltip="Staryi Kryvyn, Ukraine",
    popup="Morning of July 3",
    icon=folium.Icon(color="green", icon="train", prefix="fa"),
).add_to(m)
# --- Sytne, Ukraine -> Staryi Kryvyn, Ukraine ---
folium.PolyLine(
    locations=[sytne_coords, staryi_kryvyn_coords],
    tooltip="Sytne, Ukraine -> Staryi Kryvyn, Ukraine",
    color="red",
    weight=3,
    opacity=0.8,
    dash_array="5, 10"   # 5 px line, 10 px gap
).add_to(m)

# --- Shepetivka, Ukraine ---
shepetivka_coords = [50.1980, 27.0684]
folium.Marker(
    location=shepetivka_coords,
    tooltip="Shepetivka, Ukraine",
    popup="Morning of July 3",
    icon=folium.Icon(color="green", icon="train", prefix="fa"),
).add_to(m)
# --- Staryi Kryvyn, Ukraine -> Shepetivka, Ukraine ---
folium.PolyLine(
    locations=[staryi_kryvyn_coords, shepetivka_coords],
    tooltip="Staryi Kryvyn, Ukraine -> Shepetivka, Ukraine",
    color="red",
    weight=3,
    opacity=0.8,
    dash_array="5, 10"   # 5 px line, 10 px gap
).add_to(m)

# --- Blydni, Ukraine ---
blydni_coords = [50.1107, 27.5813]
folium.Marker(
    location=blydni_coords,
    tooltip="Blydni, Ukraine",
    popup="Morning of July 3",
    icon=folium.Icon(color="green", icon="train", prefix="fa"),
).add_to(m)
# --- Shepetivka, Ukraine -> Blydni, Ukraine ---
folium.PolyLine(
    locations=[shepetivka_coords, blydni_coords],
    tooltip="Shepetivka, Ukraine -> Blydni, Ukraine",
    color="red",
    weight=3,
    opacity=0.8,
    dash_array="5, 10"   # 5 px line, 10 px gap
).add_to(m)

# --- Pechanivka, Ukraine ---
pechanivka_coords = [50.0758, 27.8038]
folium.Marker(
    location=pechanivka_coords,
    tooltip="Pechanivka, Ukraine",
    popup="Morning of July 3",
    icon=folium.Icon(color="green", icon="train", prefix="fa"),
).add_to(m)
# --- Blydni, Ukraine -> Pechanivka, Ukraine ---
folium.PolyLine(
    locations=[blydni_coords, pechanivka_coords],
    tooltip="Blydni, Ukraine -> Pechanivka, Ukraine",
    color="red",
    weight=3,
    opacity=0.8,
    dash_array="5, 10"   # 5 px line, 10 px gap
).add_to(m)

# --- Berdychiv, Ukraine ---
berdychiv_coords = [49.8898, 28.6129]
folium.Marker(
    location=berdychiv_coords,
    tooltip="Berdychiv, Ukraine",
    popup="Morning of July 3",
    icon=folium.Icon(color="green", icon="train", prefix="fa"),
).add_to(m)
# --- Pechanivka, Ukraine -> Berdychiv, Ukraine ---
folium.PolyLine(
    locations=[pechanivka_coords, berdychiv_coords],
    tooltip="Pechanivka, Ukraine -> Berdychiv, Ukraine",
    color="red",
    weight=3,
    opacity=0.8,
    dash_array="5, 10"   # 5 px line, 10 px gap
).add_to(m)

# --- Kalynivka, Ukraine ---
kalynivka_coords = [50.2353, 30.2498]
folium.Marker(
    location=kalynivka_coords,
    tooltip="Kalynivka, Ukraine",
    popup="Afternoon of July 3",
    icon=folium.Icon(color="green", icon="train", prefix="fa"),
).add_to(m)
# --- Berdychiv, Ukraine -> Kalynivka, Ukraine ---
folium.PolyLine(
    locations=[berdychiv_coords, kalynivka_coords],
    tooltip="Berdychiv, Ukraine -> Kalynivka, Ukraine",
    color="red",
    weight=3,
    opacity=0.8,
    dash_array="5, 10"   # 5 px line, 10 px gap
).add_to(m)

# --- Kyiv-Pasazhyrskyi ---
kyiv_pas_coords = [50.4404, 30.4898]
folium.Marker(
    location=kyiv_pas_coords,
    tooltip="Kyiv-Pasazhyrskyi",
    popup="Afternoon of July 3",
    icon=folium.Icon(color="green", icon="person-running", prefix="fa"),
).add_to(m)
# --- Kalynivka, Ukraine -> Kyiv-Pasazhyrskyi ---
folium.PolyLine(
    locations=[kalynivka_coords, kyiv_pas_coords],
    tooltip="Kalynivka, Ukraine -> Kyiv-Pasazhyrskyi",
    color="red",
    weight=3,
    opacity=0.8,
    dash_array="5, 10"   # 5 px line, 10 px gap
).add_to(m)

# --- suspected residence ---
residence_coords = [50.4964, 30.5208]
folium.Marker(
    location=residence_coords,
    tooltip="Suspected Residential Address",
    popup="July 3",
    icon=folium.Icon(color="orange", icon="home"),
).add_to(m)
# uncertainty radius
folium.Circle(
    location=residence_coords,
    radius=500,
    color="orange",
    weight=2,
    fill=True,
    fill_color="orange",
    fill_opacity=0.2
).add_to(m)
# --- Kyiv-Pasazhyrskyi -> suspected residence ---
folium.PolyLine(
    locations=[kyiv_pas_coords, residence_coords],
    tooltip="Kyiv-Pasazhyrskyi -> Suspected Residence",
    color="red",
    weight=3,
    opacity=0.8,
).add_to(m)

m
```
