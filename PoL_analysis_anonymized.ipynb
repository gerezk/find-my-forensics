# Pattern-of-life (PoL) Analysis

The purpose of this notebook is to demonstrate and visualize a case study in applying techniques for pattern-of-life analysis of a target given Find My location data. The specific case is that someone stole a pair of Airpods 4 with active noice cancellation (ANC). An interesting feature of this model is that it has built-in case speakers, enabling someone on the Find My app to play a sound while the airpods are in the case. Exact addresses and some routes taken, along with modes of transportation, are known, but they have been anonymized to comply with privacy laws.

## June 28

Theft occurs on a train at approximately 11:30, most likely at Wien Hbf. The theft went unnoticed for about two hours before the Find My app was checked for the first time, where the AirPods were then located at a residential address in the 16th district (Ottakring). Therefore, the route taken by the thief to the address is unknown.

```python
m = folium.Map(location=(48.2085, 16.3721),
               zoom_start=13)

# --- location of theft ----
theft_coords = [48.1848, 16.3782]
folium.Marker(
    location=theft_coords,
    tooltip="Site of Theft",
    popup="Wien Hbf",
    icon=folium.Icon(color="red", icon="flag"),
).add_to(m)

# --- suspected residential address in Ottakring ---
ottakring_coords = [48.2128, 16.3178]
folium.Marker(
    location=ottakring_coords,
    tooltip="Suspected Residential Address",
    icon=folium.Icon(color="orange", icon="home"),
).add_to(m)
# uncertainty radius
folium.Circle(
    location=ottakring_coords,
    radius=500,
    color="orange",
    weight=2,
    fill=True,
    fill_color="orange",
    fill_opacity=0.2
).add_to(m)

# --- add unknown route ---
folium.PolyLine(
    locations=[theft_coords, ottakring_coords],
    color="red",
    weight=3,
    opacity=0.8,
    dash_array="5, 10"   # 5 px line, 10 px gap
).add_to(m)

m
```
