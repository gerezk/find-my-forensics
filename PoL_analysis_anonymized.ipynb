# Pattern-of-life (PoL) Analysis

The purpose of this notebook is to demonstrate and visualize a case study in applying techniques for pattern-of-life analysis of a target given Find My location data. The specific case is that someone stole a pair of Airpods 4 with active noice cancellation (ANC). An interesting feature of this model is that it has built-in case speakers, enabling someone on the Find My app to play a sound while the airpods are in the case. Exact information on addresses and some routes taken, along with modes of transportation, are known, but they have been anonymized to comply with privacy laws.

### Limitations

- Find My does not provide historical location data; a user must persistently monitor the app to try and build this location history. Some locations or movements could then be missing.

### Interpreting the map

- Hover over an element to get a description.
- Click on an element to show the date or time indicating when the AirPods were at a location or when travel occurred.
- A dotted line indicates that the route is unknown; in contrast, a solid line symbolizes that all route details are known.

## June 28

Theft occured on a train at approximately 11:30, most likely at Wien Hbf. The theft went unnoticed for about two hours before the Find My app was checked for the first time, where the AirPods were then located at a residential address in the 16th district (Ottakring). Therefore, the route taken by the thief to the address is unknown.

```python
m = folium.Map(location=(48.2085, 16.3721),
               zoom_start=13)

# --- location of theft ----
theft_coords = [48.1848, 16.3782]
folium.Marker(
    location=theft_coords,
    tooltip="Site of Theft - Wien Hbf",
    popup="~11:30 at June 28",
    icon=folium.Icon(color="red", icon="flag"),
).add_to(m)

# --- suspected residential address in Ottakring ---
ottakring_coords = [48.2128, 16.3178]
folium.Marker(
    location=ottakring_coords,
    tooltip="Suspected Residential Address",
    popup="June 28-29",
    icon=folium.Icon(color="orange", icon="home"),
).add_to(m)
# uncertainty radius
folium.Circle(
    location=ottakring_coords,
    radius=500,
    color="orange",
    weight=2,
    fill=True,
    fill_color="orange",
    fill_opacity=0.2
).add_to(m)

# --- add unknown route ---
folium.PolyLine(
    locations=[theft_coords, ottakring_coords],
    tooltip="Unknown route",
    popup="Afternoon of June 28",
    color="red",
    weight=3,
    opacity=0.8,
    dash_array="5, 10"   # 5 px line, 10 px gap
).add_to(m)

m
```

## June 29-30

AirPods were observed for a short time in the morning along Thaliastraße near the residential address (not marked), and police did a quick search of the area with no results, which was not surprising. The AirPods spent the rest of the day at the residential address.

Movement was detected late at night, with the target moving from Ottakring to another residential address near the Naschmarkt. The target took out the airpods from the case for the first time, and I immediately executed the "Play Sound" command on both of them. The command might have been successful since no location data was reported for the next ~48 hours, suggesting that the target reacted and did something to suppress location reports.

The exact route, timestamps, and most likely modes of transportation from the residential address in Ottakring to the Naschmarkt are known.

```python
# --- Naschmarkt address ---
naschmarkt_coords = [48.1986, 16.3612]
folium.Marker(
    location=naschmarkt_coords,
    tooltip="Airpods taken out of case for first time",
    popup="Around midnight of June 30",
    icon=folium.Icon(color="blue", icon="flash"),
).add_to(m)
# uncertainty radius
folium.Circle(
    location=naschmarkt_coords,
    radius=500,
    color="blue",
    weight=2,
    fill=True,
    fill_color="blue",
    fill_opacity=0.2
).add_to(m)

# --- add known route ---
folium.PolyLine(
    locations=[ottakring_coords, naschmarkt_coords],
    tooltip="Known route",
    popup="Around midnight of June 30",
    color="red",
    weight=3,
    opacity=0.8,
).add_to(m)

m
```
